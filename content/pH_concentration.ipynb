# pH and H+ Ion Concentration

Calculate hydrogen ion (H+ concentration for different pH values using the formula H+ = 10^(-pH). Visualize the exponential relationship between pH and H⁺ concentration.

````python3
print(" PROGRAM NAME : pH and H+ Ion Concentration Plot ")
print()

# The program calculates H+ ion concentration for different pH values.
# It provides visualization using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt

# ===================================
# DATA SETUP
# ===================================
loc = 10e-6  # Location of text for shading code block.
pH = np.array([0.8, 2.7, 4, 7, 8, 14])

# ===================================
# CALCULATION OF H+ ION CONCENTRATION
# ===================================
# Formula: pH = log_10(1/[H+]) or [H+] = 10^(-pH)
#
# The equation H_ions = 10**(-pH) calculates the hydrogen ion molar 
# concentration (H+ ions) in moles per liter (M) from a given pH value.
#
# Key Concept: As pH increases (becomes less acidic),
# the concentration of H+ ions DECREASES exponentially.
#
# Examples:
# - pH = 3 → [H+] = 10^(-3) = 1x10^-3 M (acidic)
# - pH = 7 -> [H+] = 10^(-7) = 1x10^-7 M (neutral)
# - pH = 14 -> [H+] = 10^(-14) = 1x10^-14 M (alkaline)

H_ions = 10**(-pH)  # Calculate H+ concentration

print("pH Values and Corresponding H+ Ion Concentration:")
print("="*50)
for ph_val, h_ion in zip(pH, H_ions):
    print(f"pH = {ph_val:6.1f}  →  [H⁺] = {h_ion:.2e} Mol/L")
print("="*50)
print()

# ===================================
# CREATE VISUALIZATION
# ===================================
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the main line
plt.plot(pH, H_ions, marker='o', linestyle='-', linewidth=2, 
         markersize=8, color='darkblue', label='H⁺ Concentration')

# Add scatter points for emphasis
plt.scatter(pH, H_ions, s=100, color='red', zorder=5, alpha=0.6)

# ===================================
# DATA POINT ANNOTATION
# ===================================
for x, y in zip(pH, H_ions):
    ax.annotate(f'({x}, {y:.1e})',
                xy=(x, y),
                xytext=(x + 0.5, y * 2),
                fontsize=9,
                arrowprops=dict(arrowstyle='->', color='gray', lw=1))

# ===================================
# pH RANGE SHADING (Acidic vs Alkaline)
# ===================================
ax.axvspan(0, 7, alpha=0.1, color='red', label='Acidic Region')
ax.axvspan(7, 14, alpha=0.1, color='blue', label='Alkaline Region')
ax.axvline(x=7, color='green', linestyle='--', linewidth=2, alpha=0.7, label='Neutral (pH=7)')

# Add text labels for regions
ax.text(3, loc*0.0001, "ACIDIC", color='red', fontsize=12, fontweight='bold')
ax.text(10, loc*0.0001, "ALKALINE", color='blue', fontsize=12, fontweight='bold')

# ===================================
# LABELS AND STYLING
# ===================================
plt.xlabel("pH Value", fontsize=12, fontweight='bold')
plt.ylabel("H⁺ Ion Concentration (Mol/L)", fontsize=12, fontweight='bold')
plt.title("H⁺ Ion Concentration vs pH Value\n(Logarithmic Scale)", 
          fontsize=14, fontweight='bold', pad=20)
plt.yscale('log')  # Use logarithmic scale for y-axis
plt.legend(fontsize=10, loc='upper right')
plt.grid(True, alpha=0.3, linestyle='--')

# ===================================
# ADD FORMULA
# ===================================
equation = r'$[H^+] = 10^{-pH}$'
ax.text(0.5, loc*0.005, equation, fontsize=13, 
        bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))

plt.tight_layout()
plt.show()

print("\n✅ Graph generated successfully!")
````

## 🧪 Key Chemistry Concepts

### **pH Scale Definition**
- **pH = -log_10[H+]** or **[H+] = 10^(-pH)**
- Scale ranges from 0 (very acidic) to 14 (very alkaline)
- pH = 7 is neutral (pure water at 25°C)

### **Regions:**
- **Acidic:** pH < 7 (high H+ concentration)
- **Neutral:** pH = 7 (10^-7 = 0.0000001 M)
- **Alkaline (Basic):** pH > 7 (low H⁺ concentration)

### **Exponential Relationship**
Notice the **logarithmic scale** on the y-axis! This shows how dramatically the H+ concentration changes:
- A **1-unit decrease in pH** = **10x increase** in H+ concentration
- Example: pH 3 has 10x more H+ ions than pH 4
