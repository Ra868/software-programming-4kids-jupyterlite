# Python Tutorial with NumPy & Pandas

A quick introduction to scientific computing in JupyterLite.

## 1. Working with NumPy Arrays

```python
import numpy as np

# Create arrays
arr = np.array([1, 2, 3, 4, 5])
print('Array:', arr)
print('Mean:', np.mean(arr))
print('Std Dev:', np.std(arr))
print('Sum:', np.sum(arr))
```

## 2. Data Analysis with Pandas

```python
import pandas as pd

# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Score': [85, 90, 78, 92]
}

df = pd.DataFrame(data)
print(df)
print('\nSummary Statistics:')
print(df.describe())
```

## 3. Data Visualization

```python
import matplotlib.pyplot as plt

# Bar chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Ages
ax1.bar(df['Name'], df['Age'], color='skyblue', edgecolor='navy')
ax1.set_title('Age Distribution')
ax1.set_ylabel('Age')
ax1.grid(axis='y', alpha=0.3)

# Scores
ax2.scatter(df['Age'], df['Score'], s=100, color='coral', edgecolor='darkred')
ax2.set_title('Age vs Score')
ax2.set_xlabel('Age')
ax2.set_ylabel('Score')
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
