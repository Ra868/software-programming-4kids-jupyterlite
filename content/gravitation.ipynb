## 🔬 The Physics Behind the Formula

### **Newton's Universal Law of Gravitation**

$$g = \frac{GM}{r^2}$$

| Symbol | Meaning | Value / Unit |
|--------|---------|-------------|
| **g** | Gravitational acceleration | m/s² |
| **G** | Universal gravitational constant | 6.674 × 10⁻¹¹ N·m²/kg² |
| **M** | Mass of Earth | 5.972 × 10²⁴ kg |
| **r** | Distance from Earth's center | meters |

### **Key Insight: Inverse-Square Law**

Gravity follows an **inverse-square relationship** with distance:
- Double the distance → **¼** of the gravity
- Triple the distance → **1/9** of the gravity
- 10× the distance → **1/100** of the gravity

### **Real-World Context**

- 🛫 **Commercial flights** cruise at ~10–12 km altitude — gravity there is nearly identical to the surface!
- 🛸 **International Space Station** orbits at ~400 km — gravity is still ~88% of surface value!
- 🌙 **Moon's orbit** is ~384,400 km away — gravity from Earth is only ~0.03% of surface value there.

````python3
import numpy as np

# =================================
# CONSTANTS (self-contained cell)
# =================================
G = 6.674e-11
M_earth = 5.972e24
R_earth = 6.371e6
R_moon_orbit   = 3.844e8
R_moon_surface = 1.737e6

locations = [
    ("Earth's Surface",                          R_earth),
    ("10 km altitude",                           R_earth + 1.0e4),
    ("15 km altitude",                           R_earth + 1.5e4),
    ("20 km altitude",                           R_earth + 2.0e4),
    ("100,000 km altitude",                      R_earth + 1.0e8),
    ("200,000 km altitude",                      R_earth + 2.0e8),
    ("300,000 km altitude",                      R_earth + 3.0e8),
    ("Moon's Surface (at Moon's orbital dist.)", R_moon_orbit + R_moon_surface),
]

# =================================
# CALCULATE GRAVITATIONAL ACCELERATION
# =================================
distances_m = np.array([loc[1] for loc in locations])
distances_km = distances_m / 1e3
distances_from_surface_km = (distances_m - R_earth) / 1e3

# g = GM / r^2
g_values = G * M_earth / distances_m**2
g_surface = g_values[0]  # reference value at Earth's surface

# =================================
# PRINT RESULTS TABLE
# =================================
print("Gravitational Acceleration — Newton's Law: g = GM/r²")
print("=" * 106)
print(f"{'Location':<44} {'Distance from center (km)':>25} {'Altitude (km)':>14} {'g (m/s²)':>12} {'g / g_surface':>13}")
print("-" * 106)
for i, (label, r_m) in enumerate(locations):
    print(f"{label:<44} {distances_km[i]:>25,.1f} {distances_from_surface_km[i]:>14.1f} "
          f"{g_values[i]:>12.6f} {g_values[i]/g_surface:>13.6f}")
print("=" * 106)
print(f"\nSurface gravity: g = {g_surface:.4f} m/s²")
print("[OK] Calculations complete!")
````

````python3
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# =================================
# CONSTANTS (self-contained cell)
# =================================
G = 6.674e-11
M_earth = 5.972e24
R_earth = 6.371e6
R_moon_orbit   = 3.844e8
R_moon_surface = 1.737e6

locations = [
    ("Earth's Surface",                          R_earth),
    ("10 km altitude",                           R_earth + 1.0e4),
    ("15 km altitude",                           R_earth + 1.5e4),
    ("20 km altitude",                           R_earth + 2.0e4),
    ("100,000 km altitude",                      R_earth + 1.0e8),
    ("200,000 km altitude",                      R_earth + 2.0e8),
    ("300,000 km altitude",                      R_earth + 3.0e8),
    ("Moon's Surface (at Moon's orbital dist.)", R_moon_orbit + R_moon_surface),
]

colors = ["#3fb950", "#ffa657", "#e3b341", "#f78166",
          "#d2a8ff", "#79c0ff", "#58a6ff", "#ff7b72"]

distances_m = np.array([loc[1] for loc in locations])
distances_km = distances_m / 1e3
distances_from_surface_km = (distances_m - R_earth) / 1e3
g_values = G * M_earth / distances_m**2
g_surface = g_values[0]

# Smooth curve arrays
r_plot = np.linspace(R_earth, R_moon_orbit + R_moon_surface + 5e6, 2000)
g_plot = G * M_earth / r_plot**2
r_plot_km = r_plot / 1e3

# =================================
# FIGURE LAYOUT (dark theme)
# =================================
fig = plt.figure(figsize=(16, 11), facecolor="#0d1117")
fig.suptitle(
    "Acceleration Due to Earth's Gravity — Newton's Universal Law of Gravitation  |  g = GM/r²",
    fontsize=14, fontweight="bold", color="white", y=0.99,
)

gs = gridspec.GridSpec(
    2, 2,
    height_ratios=[2.0, 1.15],
    width_ratios=[1.6, 1],
    hspace=0.42, wspace=0.32,
    left=0.07, right=0.97, top=0.94, bottom=0.04,
)

# ── Main plot: full scale (Surface → Moon) ────────────────────────────────
ax_main = fig.add_subplot(gs[0, 0])
ax_main.set_facecolor("#161b22")
ax_main.plot(r_plot_km / 1e3, g_plot,
             color="#58a6ff", linewidth=2.5, zorder=2, label="g(r)")
ax_main.set_title("Full Scale  (Surface → Moon)", color="#8b949e", fontsize=10, pad=6)

far_indices = [0, 4, 5, 6, 7]
for i in far_indices:
    label, r_m = locations[i]
    r_km = r_m / 1e3
    g = g_values[i]
    ax_main.scatter(r_km / 1e3, g, color=colors[i], s=90, zorder=5)
    offset_y = g + 0.06 * g_surface * max(0.05, 1 - far_indices.index(i) * 0.18)
    ax_main.annotate(
        label, xy=(r_km / 1e3, g),
        xytext=(r_km / 1e3 + 8, offset_y),
        color=colors[i], fontsize=8,
        arrowprops=dict(arrowstyle="->", color=colors[i], lw=0.9), zorder=6,
    )

ax_main.set_xlabel("Distance from Earth's Center  (×10³ km)", color="#8b949e", fontsize=10)
ax_main.set_ylabel("Gravitational Acceleration  (m/s²)", color="#8b949e", fontsize=10)
ax_main.tick_params(colors="#8b949e", labelsize=8.5)
for sp in ax_main.spines.values():
    sp.set_edgecolor("#30363d")
ax_main.grid(True, color="#21262d", linestyle="--", linewidth=0.8, zorder=1)
ax_main.set_xlim(left=0)
ax_main.set_ylim(bottom=0)
ax_main.legend(facecolor="#161b22", edgecolor="#30363d", labelcolor="white", fontsize=8.5)

# ── Zoomed plot: near-Earth (Surface → 50 km altitude) ───────────────────
ax_zoom = fig.add_subplot(gs[0, 1])
ax_zoom.set_facecolor("#161b22")

r_zoom = np.linspace(R_earth, R_earth + 6e4, 500)
g_zoom = G * M_earth / r_zoom**2
r_zoom_alt_km = (r_zoom - R_earth) / 1e3

ax_zoom.plot(r_zoom_alt_km, g_zoom, color="#58a6ff", linewidth=2.5, zorder=2)
ax_zoom.set_title("Zoomed  (Surface to 50 km altitude)", color="#8b949e", fontsize=10, pad=6)

near_indices = [0, 1, 2, 3]
for i in near_indices:
    label, r_m = locations[i]
    alt_km = (r_m - R_earth) / 1e3
    g = g_values[i]
    ax_zoom.scatter(alt_km, g, color=colors[i], s=90, zorder=5)
    ax_zoom.annotate(
        label, xy=(alt_km, g),
        xytext=(alt_km + 1.5, g - 0.0008 * (near_indices.index(i) + 1)),
        color=colors[i], fontsize=8.5,
        arrowprops=dict(arrowstyle="->", color=colors[i], lw=0.9), zorder=6,
    )

ax_zoom.set_xlabel("Altitude above Earth's Surface  (km)", color="#8b949e", fontsize=10)
ax_zoom.set_ylabel("Gravitational Acceleration  (m/s²)", color="#8b949e", fontsize=10)
ax_zoom.tick_params(colors="#8b949e", labelsize=8.5)
for sp in ax_zoom.spines.values():
    sp.set_edgecolor("#30363d")
ax_zoom.grid(True, color="#21262d", linestyle="--", linewidth=0.8, zorder=1)
ax_zoom.set_xlim(-1, 52)
ax_zoom.set_ylim(g_zoom[-1] - 0.001, g_zoom[0] + 0.003)

# ── Data table (spans both columns) ──────────────────────────────────────
ax_table = fig.add_subplot(gs[1, :])
ax_table.set_facecolor("#161b22")
ax_table.axis("off")

col_labels = [
    "Location",
    "Distance from\nEarth's Center (km)",
    "Altitude above\nEarth's Surface (km)",
    "Gravitational\nAcceleration (m/s²)",
    "Fraction of\nSurface g",
]

table_data = []
for i, (label, r_m) in enumerate(locations):
    dist_str = f"{distances_km[i]:,.3f}" if distances_from_surface_km[i] < 1 else f"{distances_km[i]:,.0f}"
    alt_str  = f"{distances_from_surface_km[i]:,.1f}" if distances_from_surface_km[i] < 100 else f"{distances_from_surface_km[i]:,.0f}"
    table_data.append([label, dist_str, alt_str,
                       f"{g_values[i]:.6f}", f"{g_values[i]/g_surface:.6f}"])

table = ax_table.table(
    cellText=table_data, colLabels=col_labels,
    cellLoc="center", loc="center", bbox=[0, 0, 1, 1],
)
table.auto_set_font_size(False)
table.set_fontsize(8.5)

header_color = "#1f6feb"
row_bg = ["#161b22", "#1c2128"]

for (row, col), cell in table.get_celld().items():
    cell.set_edgecolor("#30363d")
    if row == 0:
        cell.set_facecolor(header_color)
        cell.set_text_props(color="white", fontweight="bold")
    else:
        cell.set_facecolor(row_bg[(row - 1) % 2])
        if col == 0:
            cell.set_text_props(color=colors[row - 1], fontweight="bold")
        else:
            cell.set_text_props(color="#c9d1d9")

plt.show()
print("[OK] Visualization complete!")
````

## 💡 Key Insights

### **What the Data Tells Us**

| Location | g (m/s²) | % of Surface g |
|----------|----------|----------------|
| Earth's Surface | 9.82 | 100% |
| 10 km altitude | 9.79 | ~99.7% |
| 20 km altitude | 9.76 | ~99.4% |
| ISS (~400 km) | ~8.69 | ~88% |
| 100,000 km altitude | ~0.03 | ~0.3% |
| Moon's orbit | ~0.003 | ~0.03% |

### **Surprising Facts 🤯**

1. **Airplanes feel nearly full gravity!** At 10 km altitude, g is only 0.3% less than at sea level.
2. **Astronauts on the ISS are NOT weightless!** Gravity there is ~88% of surface value — they're in *free fall*, not zero gravity.
3. **Gravity never reaches zero** — it decreases forever following the 1/r² law but never becomes exactly zero.
4. **The Moon stays in orbit** because Earth's gravity at that distance (0.003 m/s²) is just enough to keep it moving in a circle.

### **How Gravity Decreases**

The inverse-square law means:
- Moving from surface (6,371 km) to Moon's orbit (384,400 km) increases distance by **60×**
- Gravity decreases by **60² = 3,600×**
- Surface g ≈ 9.82 m/s², Moon's orbit g ≈ 0.0027 m/s² (about 3,600 times smaller ✅)

````python3
import numpy as np
import math

# =================================
# CONSTANTS
# =================================
G = 6.674e-11
M_earth = 5.972e24
R_earth = 6.371e6

# =================================
# EXPERIMENT 2: Gravity on Other Planets
# =================================
print("Surface Gravity on Solar System Bodies")
print("=" * 45)

bodies = {
    "Earth":   {"M": 5.972e24, "R": 6.371e6},
    "Moon":    {"M": 7.342e22, "R": 1.737e6},
    "Mars":    {"M": 6.390e23, "R": 3.390e6},
    "Venus":   {"M": 4.867e24, "R": 6.052e6},
    "Jupiter": {"M": 1.898e27, "R": 6.991e7},
    "Saturn":  {"M": 5.683e26, "R": 5.823e7},
}

g_earth = G * bodies["Earth"]["M"] / bodies["Earth"]["R"]**2

for name, data in bodies.items():
    g_body = G * data["M"] / data["R"]**2
    ratio  = g_body / g_earth
    print(f"{name:<10}  g = {g_body:6.2f} m/s²  ({ratio:.2f}× Earth)")

# =================================
# EXPERIMENT 3: Escape Velocity from Earth
# =================================
print()
print("Escape Velocities")
print("=" * 45)
for name, data in bodies.items():
    v_esc = math.sqrt(2 * G * data["M"] / data["R"])
    print(f"{name:<10}  v_escape = {v_esc/1000:7.2f} km/s")

print()
print("[OK] Experiments complete! Try modifying the values above.")
````
