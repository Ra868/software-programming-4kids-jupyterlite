````python3
print("PROGRAM NAME: Atmospheric Pressure vs Altitude Analysis\n")

import numpy as np
import matplotlib.pyplot as plt

# ==============================================
# BAROMETRIC FORMULA
# ==============================================
# P(h) = P₀ × e^(-Mgh/RT)
#
# Where:
# P(h) = Atmospheric pressure at altitude h
# P₀ = Pressure at sea level (101,325 Pa)
# M = Molar mass of Earth's air (0.029 kg/mol)
# g = Acceleration due to gravity (9.81 m/s²)
# h = Altitude (meters)
# R = Universal gas constant (8.314 J/mol·K)
# T = Temperature in Kelvin (288 K ≈ 15°C at sea level)
#
# Key Concept: Pressure decreases exponentially with altitude
# because the atmosphere has less mass above you.

# ==============================================
# CONSTANTS
# ==============================================
Po = 101325      # Atmospheric pressure at sea level (Pascal)
M = 0.029        # Molar mass of Earth's air (kg/mol)
g = 9.81         # Acceleration due to gravity (m/s²)
R = 8.314        # Universal gas constant (J/mol·K)
T = 288          # Temperature in Kelvin (≈ 15°C)

# Altitudes to analyze (meters)
h = np.array([0, 500, 1000, 2000, 5000, 11000, 15000])

# ==============================================
# CALCULATE ATMOSPHERIC PRESSURE
# ==============================================
# Using the barometric formula: P = P₀ × e^(-Mgh/RT)
p = Po * np.exp(-M * g * h / (R * T))

# Round to nearest Pascal
p = np.round(p, 0)

print("Atmospheric Pressure at Different Altitudes")
print("="*70)
print(f"Sea Level Reference: P₀ = {Po:,} Pa = {Po/101325:.2f} atm")
print(f"Temperature: {T} K ({T-273.15:.1f}°C)")
print("="*70)
print()

# Convert to other units for reference
p_atm = p / 101325  # Convert to atmospheres
p_bar = p / 100000  # Convert to bars

print(f"{'Altitude (m)':>15} {'Pressure (Pa)':>20} {'Pressure (atm)':>18} {'% of Sea Level':>18}")
print("-"*70)
for altitude, pressure, pressure_atm in zip(h, p, p_atm):
    percent = (pressure / Po) * 100
    print(f"{int(altitude):>15,} {int(pressure):>20,} {pressure_atm:>18.4f} {percent:>17.1f}%")
print("-"*70)
````

## 📊 Data Table Visualization

````python3
# Create a nicely formatted table
row_labels = [
    'Sea Level (0 m)',
    'Altitude = 500 m',
    'Altitude = 1,000 m',
    'Altitude = 2,000 m',
    'Altitude = 5,000 m',
    'Troposphere Top (11,000 m)',
    'Altitude = 15,000 m'
]

col_labels = ['Pressure (Pa)', 'Pressure (atm)', 'Percent of P₀']

# Prepare table data
table_data = []
for pressure, pressure_atm in zip(p, p_atm):
    percent = (pressure / Po) * 100
    table_data.append([
        f"{int(pressure):,}",
        f"{pressure_atm:.4f}",
        f"{percent:.1f}%"
    ])

fig, ax = plt.subplots(figsize=(12, 6))
ax.axis('off')

table = ax.table(
    cellText=table_data,
    rowLabels=row_labels,
    colLabels=col_labels,
    loc='center',
    cellLoc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 2.2)

# Style the header
for i in range(len(col_labels)):
    table[(0, i)].set_facecolor('#4472C4')
    table[(0, i)].set_text_props(weight='bold', color='white')

# Alternate row colors
for i in range(1, len(row_labels) + 1):
    if i % 2 == 0:
        for j in range(len(col_labels)):
            table[(i, j)].set_facecolor('#E7E6E6')

plt.title('Atmospheric Pressure Data Table', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
````

## 📈 Pressure vs Altitude Graph (Linear Scale)

````python3
fig, ax = plt.subplots(figsize=(11, 7))

# Plot main line and points
ax.plot(h, p, marker='o', linestyle='-', linewidth=2.5, 
        markersize=10, color='#2E75B6', alpha=0.8, label='Atmospheric Pressure')
ax.scatter(h, p, s=150, color='red', zorder=5, alpha=0.6, edgecolors='darkred', linewidth=2)

# ===== ATMOSPHERIC LAYER SHADING =====
# Troposphere (0-11,000 m)
ax.axvspan(0, 11000, alpha=0.1, color='blue', label='Troposphere')
ax.text(5500, 55000, 'TROPOSPHERE\n(0-11 km)', fontsize=12, 
        fontweight='bold', ha='center',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7))

# Stratosphere (11,000-50,000 m)
ax.axvspan(11000, 15000, alpha=0.1, color='red', label='Stratosphere')
ax.text(13000, 20000, 'STRATOSPHERE\n(11-50 km)', fontsize=12, 
        fontweight='bold', ha='center', color='darkred',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.7))

# ===== DATA POINT ANNOTATIONS =====
for x, y in zip(h, p):
    ax.annotate(f'{int(x):,}m\n{int(y):,}Pa',
                xy=(x, y),
                xytext=(x + 400, y + 5000),
                fontsize=9,
                arrowprops=dict(arrowstyle='->', color='gray', lw=1.5),
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.5))

# ===== SECONDARY Y-AXIS (ATM) =====
def pa_to_atm(pa):
    return pa / 101325

def atm_to_pa(atm):
    return atm * 101325

secax = ax.secondary_yaxis('right', functions=(pa_to_atm, atm_to_pa))
secax.set_ylabel("Atmospheric Pressure (atm)", fontsize=12, fontweight='bold')

# ===== BAROMETRIC FORMULA =====
equation = r'$P(h) = P_0 e^{-\frac{Mgh}{RT}}$'
ax.text(0.02, 0.95, equation, transform=ax.transAxes, fontsize=14,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

# ===== LABELS AND STYLING =====
ax.set_xlabel("Altitude (meters)", fontsize=12, fontweight='bold')
ax.set_ylabel("Atmospheric Pressure (Pascal)", fontsize=12, fontweight='bold')
ax.set_title("Atmospheric Pressure vs Altitude (Linear Scale)", 
             fontsize=14, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(fontsize=11, loc='upper right', framealpha=0.95)

plt.tight_layout()
plt.show()
````

## 📊 Pressure vs Altitude Graph (Logarithmic Scale)

````python3
fig, ax = plt.subplots(figsize=(11, 7))

# Plot main line and points
ax.plot(h, p, marker='o', linestyle='-', linewidth=2.5, 
        markersize=10, color='#2E75B6', alpha=0.8, label='Atmospheric Pressure')
ax.scatter(h, p, s=150, color='red', zorder=5, alpha=0.6, edgecolors='darkred', linewidth=2)

# ===== ATMOSPHERIC LAYER SHADING =====
ax.axvspan(0, 11000, alpha=0.1, color='blue')
ax.axvspan(11000, 15000, alpha=0.1, color='red')

# ===== DATA POINT ANNOTATIONS =====
for x, y in zip(h, p):
    ax.annotate(f'{int(x):,}m',
                xy=(x, y),
                xytext=(x + 400, y * 1.3),
                fontsize=9,
                arrowprops=dict(arrowstyle='->', color='gray', lw=1.5))

# ===== SECONDARY Y-AXIS (ATM) =====
secax = ax.secondary_yaxis('right', functions=(pa_to_atm, atm_to_pa))
secax.set_ylabel("Atmospheric Pressure (atm)", fontsize=12, fontweight='bold')

# ===== USE LOGARITHMIC SCALE =====
ax.set_yscale('log')

# ===== LABELS AND STYLING =====
ax.set_xlabel("Altitude (meters)", fontsize=12, fontweight='bold')
ax.set_ylabel("Atmospheric Pressure (Pascal) [Log Scale]", fontsize=12, fontweight='bold')
ax.set_title("Atmospheric Pressure vs Altitude (Logarithmic Scale)", 
             fontsize=14, fontweight='bold', pad=20)
ax.grid(True, which='both', alpha=0.3, linestyle='--')
ax.legend(fontsize=11, loc='upper right', framealpha=0.95)

# ===== ADD NOTE =====
ax.text(0.5, 0.05, 'On a log scale, exponential decay appears as a straight line!', 
        transform=ax.transAxes, fontsize=11, ha='center',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7))

plt.tight_layout()
plt.show()
````

## 🌍 Atmospheric Science Concepts

### **The Barometric Formula**
$$P(h) = P_0 e^{-\frac{Mgh}{RT}}$$

This formula describes how **atmospheric pressure decreases exponentially with altitude** due to the decreasing mass of air above you.

### **Key Parameters**
- **P₀ = 101,325 Pa** - Standard atmospheric pressure at sea level
- **M = 0.029 kg/mol** - Average molar mass of air
- **g = 9.81 m/s²** - Gravitational acceleration
- **R = 8.314 J/(mol·K)** - Universal gas constant
- **T = 288 K (15°C)** - Standard temperature

### **Atmospheric Layers**

1. **Troposphere (0-11 km)**
   - Where we live and weather occurs
   - Temperature decreases with altitude
   - Contains 80% of atmosphere's mass
   - At 11 km: pressure ≈ 23,800 Pa (23.5% of sea level)

2. **Stratosphere (11-50 km)**
   - Temperature increases with altitude (ozone layer!)
   - Very thin air
   - At 15 km: pressure ≈ 12,100 Pa (12% of sea level)

### **Real-World Observations**
- **Mount Everest (8,849 m)**: ~33% of sea-level pressure → Low oxygen, need supplemental O₂
- **Commercial Airlines (10,000 m)**: ~26% of sea-level pressure → Pressurized cabins required
- **Death Zone (8,000 m)**: Human body cannot acclimatize above this altitude

### **Unit Conversions**
- 1 atm = 101,325 Pa
- 1 bar = 100,000 Pa ≈ 0.987 atm
- 1 psi = 6,894.76 Pa
