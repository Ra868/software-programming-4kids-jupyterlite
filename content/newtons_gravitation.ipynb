# Planet Mass Calculator

## Introduction to Newtons Universal Law of Gravitation

Newtons Universal Law of Gravitation states that the gravitational force between two masses is:

$$F = G \frac{M \cdot m}{r^2}$$

Where:
- G = 6.674 x 10^-11 N·m²/kg² (gravitational constant)
- M = mass of planet (kg)
- m = mass of object (kg)
- r = distance between centers (m)

At a planets surface, gravity g is related to planetary mass by:

$$g = G \frac{M}{R^2}$$

Solving for mass:

$$M = \frac{g \cdot R^2}{G}$$

In this lesson, we calculate the mass of planets using their surface gravity and radius.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd

# Data setup
G = 6.674e-11
names = np.array(['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter'])
surface_gravity = np.array([3.70, 8.87, 9.81, 3.71, 24.79])
radii = np.array([2.4397e6, 6.0518e6, 6.3710e6, 3.3895e6, 7.1492e7])
known_masses = np.array([3.3011e23, 4.8675e24, 5.9720e24, 6.4171e23, 1.8982e27])
colors = ['#b5b5b5', '#e8cda0', '#4a90d9', '#c1440e', '#c88b3a']

print("Data loaded")
print(f"G = {G:.3e}")
```

## Step 1: Calculate Masses

```python
# Calculate masses using M = (g * R^2) / G
calculated_masses = (surface_gravity * radii**2) / G
percent_error = np.abs(calculated_masses - known_masses) / known_masses * 100

print("Calculation complete")
```

## Step 2: Results Table

```python
results = pd.DataFrame({
    'Planet': names,
    'g (m/s^2)': surface_gravity,
    'Radius (m)': radii,
    'Calculated (kg)': calculated_masses,
    'Known (kg)': known_masses,
    'Error %': percent_error
})

print("\n" + "="*90)
print("Planet Mass Results")
print("="*90)
print(results.to_string(index=False))
print("="*90)
print(f"Largest: {names[np.argmax(calculated_masses)]}")
print(f"Smallest: {names[np.argmin(calculated_masses)]}")
print(f"Max error: {percent_error.max():.5f}%")
```

## Step 3: Visualizations

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Planet Mass Calculations", fontsize=14, fontweight='bold')

x = np.arange(len(names))

# Linear scale
ax1.bar(x, calculated_masses, color=colors, edgecolor='white', linewidth=0.8)
ax1.set_xticks(x)
ax1.set_xticklabels(names)
ax1.set_ylabel('Mass (kg)')
ax1.set_title('Linear Scale')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Log scale
ax2.bar(x, calculated_masses, color=colors, edgecolor='white', linewidth=0.8)
ax2.set_xticks(x)
ax2.set_xticklabels(names)
ax2.set_ylabel('Mass (kg)')
ax2.set_yscale('log')
ax2.set_title('Logarithmic Scale')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print("Done")
```
