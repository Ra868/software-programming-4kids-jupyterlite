````python3
print("PROGRAM NAME: Compound Interest Calculator\n")

import numpy as np
import matplotlib.pyplot as plt

# ==============================================
# COMPOUND INTEREST FORMULA
# ==============================================
# A = P(1 + r)^t
#
# Where:
# A = Final amount
# P = Principal (initial investment)
# r = Interest rate (as decimal, e.g., 0.05 for 5%)
# t = Time in years
#
# Key Concept: Your money grows EXPONENTIALLY!
# The longer you invest, the more dramatic the growth.

# ==============================================
# SETUP
# ==============================================
P = 100  # Initial Principal in dollars

# Different interest rates (annual percentage rate)
rates = np.array([0.02, 0.05, 0.06])  # 2%, 5%, 6%

# Different time periods
years = np.array([3, 6, 7])  # 3, 6, and 7 years

print(f"Initial Investment (Principal): ${P}")
print(f"Interest Rates: {rates * 100}%")
print(f"Time Periods: {years} years")
print()

# ==============================================
# CALCULATE COMPOUND INTEREST
# ==============================================
# Using NumPy broadcasting:
# rates[:, np.newaxis] creates a column vector
# This allows vectorized calculation across all combinations

# rates[:, np.newaxis] shape: (3, 1)
# years shape: (3,) → broadcasts to (3, 3)
# Result: 3x3 matrix of all rate-year combinations

amounts = P * (1 + rates[:, np.newaxis]) ** years

# Round to 2 decimal places (currency)
amounts = np.round(amounts, 2)

# ==============================================
# DISPLAY TABLE
# ==============================================
print("Final Amount for $100 Investment")
print("="*60)
print(f"{'Interest Rate':>15} {'After 3 yrs':>15} {'After 6 yrs':>15} {'After 7 yrs':>15}")
print("-"*60)

for i, rate in enumerate(rates):
    rate_pct = int(rate * 100)
    print(f"{rate_pct}%{'':<12} ${amounts[i, 0]:>14.2f} ${amounts[i, 1]:>14.2f} ${amounts[i, 2]:>14.2f}")
    # Calculate gains
    gains = amounts[i] - P
    print(f"{'Gain':>15} ${gains[0]:>14.2f} ${gains[1]:>14.2f} ${gains[2]:>14.2f}")
    print()

print("="*60)
print()
print("Key Insights:")
max_gain_idx = np.argmax(amounts)
max_gain_rate_idx = max_gain_idx // amounts.shape[1]
max_gain_year_idx = max_gain_idx % amounts.shape[1]
print(f"• Highest return: {int(rates[max_gain_rate_idx]*100)}% over {years[max_gain_year_idx]} years = ${amounts[max_gain_rate_idx, max_gain_year_idx]}")
print(f"• Time multiplier effect: 7 years vs 3 years")
for i, rate in enumerate(rates):
    multiplier = amounts[i, 2] / amounts[i, 0]
    print(f"  At {int(rate*100)}%: {multiplier:.2f}x growth (from ${amounts[i,0]:.2f} to ${amounts[i,2]:.2f})")
````

## 📊 Data Table Visualization

````python3
# Setup - define variables needed for visualization
import numpy as np
import matplotlib.pyplot as plt

P = 100
rates = np.array([0.02, 0.05, 0.06])
years = np.array([3, 6, 7])
amounts = P * (1 + rates[:, np.newaxis]) ** years
amounts = np.round(amounts, 2)

# Create a nicely formatted table
row_labels = ['2% Interest', '5% Interest', '6% Interest']
col_labels = ['After 3 Years', 'After 6 Years', 'After 7 Years']

# Prepare table data with dollar formatting
table_data = []
for i in range(amounts.shape[0]):
    row = []
    for j in range(amounts.shape[1]):
        row.append(f"${amounts[i, j]:.2f}")
    table_data.append(row)

fig, ax = plt.subplots(figsize=(12, 4))
ax.axis('off')

table = ax.table(
    cellText=table_data,
    rowLabels=row_labels,
    colLabels=col_labels,
    loc='center',
    cellLoc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 2)

# Style the header
for i in range(len(col_labels)):
    table[(0, i)].set_facecolor('#4472C4')
    table[(0, i)].set_text_props(weight='bold', color='white')

# Style row labels
for i in range(1, len(row_labels) + 1):
    table[(i, -1)].set_facecolor('#D9E1F2')
    table[(i, -1)].set_text_props(weight='bold')

# Alternate row colors
for i in range(1, len(row_labels) + 1):
    if i % 2 == 0:
        for j in range(len(col_labels)):
            table[(i, j)].set_facecolor('#E7E6E6')

plt.title('Compound Interest: Final Amount for $100 Investment', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
````

## 📈 Line Plot: Investment Growth Over Different Time Periods

````python3
fig, ax = plt.subplots(figsize=(11, 7))

# Define colors for each interest rate
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

# Plot each interest rate
for i, rate in enumerate(rates):
    rate_pct = int(rate * 100)
    ax.plot(years, amounts[i], 
            marker='o', 
            linewidth=2.5, 
            markersize=10,
            label=f'{rate_pct}% Interest',
            color=colors[i],
            alpha=0.8)

# ===== DATA POINT ANNOTATIONS =====
for i, rate in enumerate(rates):
    for j, year in enumerate(years):
        ax.annotate(f'${amounts[i, j]:.2f}',
                    xy=(year, amounts[i, j]),
                    xytext=(5, 10),
                    textcoords='offset points',
                    fontsize=9,
                    bbox=dict(boxstyle='round,pad=0.3', 
                             facecolor=colors[i], 
                             alpha=0.2))

# ===== LABELS AND STYLING =====
ax.set_xlabel("Years", fontsize=12, fontweight='bold')
ax.set_ylabel("Final Amount ($)", fontsize=12, fontweight='bold')
ax.set_title("Growth of $100 Investment at Different Interest Rates", 
             fontsize=14, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(fontsize=11, loc='upper left', framealpha=0.95)

# ===== FORMULA =====
equation = r'$A = P(1 + r)^t$'
ax.text(0.98, 0.05, equation, transform=ax.transAxes, fontsize=14,
        verticalalignment='bottom', horizontalalignment='right',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.show()
````

## 💹 Extended Analysis: More Time Periods

````python3
# Let's extend the analysis to see long-term growth
extended_years = np.arange(0, 31, 1)  # 0 to 30 years

fig, ax = plt.subplots(figsize=(12, 7))

# Calculate amounts for extended time period
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

for i, rate in enumerate(rates):
    extended_amounts = P * (1 + rate) ** extended_years
    rate_pct = int(rate * 100)
    
    ax.plot(extended_years, extended_amounts,
            linewidth=2.5,
            label=f'{rate_pct}% Interest',
            color=colors[i],
            alpha=0.8)

# ===== ADD ANNOTATIONS FOR KEY MILESTONES =====
# Show when investment doubles
for i, rate in enumerate(rates):
    # Find when amount = 2*P (doubles)
    double_time = np.log(2) / np.log(1 + rate)
    double_amount = P * (1 + rate) ** double_time
    ax.plot(double_time, double_amount, 'x', markersize=12, 
            markeredgewidth=2, color=colors[i])
    ax.annotate(f'2x at {double_time:.1f} yrs',
                xy=(double_time, double_amount),
                xytext=(10, 20),
                textcoords='offset points',
                fontsize=9,
                arrowprops=dict(arrowstyle='->', color=colors[i]))

# ===== LABELS AND STYLING =====
ax.set_xlabel("Years", fontsize=12, fontweight='bold')
ax.set_ylabel("Final Amount ($)", fontsize=12, fontweight='bold')
ax.set_title("Long-Term Investment Growth: 30-Year Projection", 
             fontsize=14, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(fontsize=11, loc='upper left', framealpha=0.95)

plt.tight_layout()
plt.show()

print("\nKey Finding - Rule of 72:")
print("To find how long it takes to double your money:")
print("Time to double ≈ 72 / interest_rate_percent")
print()
for i, rate in enumerate(rates):
    double_time = np.log(2) / np.log(1 + rate)
    rule_of_72 = 72 / (rate * 100)
    actual = P * (1 + rate) ** double_time
    print(f"At {int(rate*100)}%: Actual = {double_time:.2f} years, Rule of 72 = {rule_of_72:.2f} years")
````

## 📚 Mathematics & Finance Concepts

### **Compound Interest Formula**
$$A = P(1 + r)^t$$

Where:
- **A** = Final amount (what you have at the end)
- **P** = Principal (initial investment)
- **r** = Interest rate (as decimal: 0.05 for 5%)
- **t** = Time in years

### **Key Concepts**

1. **Exponential Growth**
   - The exponent `t` makes this an exponential function
   - Small changes in rate/time create BIG changes in final amount
   - Growth accelerates over time

2. **Time Value of Money**
   - $100 today ≠ $100 in 10 years
   - With 5% interest, $100 becomes $162.89 in 10 years
   - This is why starting early matters!

3. **Rule of 72**
   - Approximate formula: Years to double ≈ 72 / interest_rate%
   - At 5%: 72/5 = 14.4 years (very close to actual ~14.2 years)
   - At 6%: 72/6 = 12 years (very close to actual ~11.9 years)

4. **Comparison of Rates**
   - 2% vs 5% seems small (3% difference)
   - But over 7 years: $117.86 vs $140.71 = **$22.85 more** (19% better!)
   - Over 30 years: $181.14 vs $432.19 = **$251 more** (138% better!)

### **Real-World Examples**

- **Savings Account:** 0.5% - 2% (very low growth)
- **CD (Certificate of Deposit):** 4% - 5%
- **Stock Market Average:** ~10% historically (variable, risky)
- **Inflation Rate:** ~3% (your money loses value if returns < inflation)

### **Continuous Compounding**
When interest compounds continuously (not just yearly):
$$A = Pe^{rt}$$

This is slightly more generous, especially for high rates!
