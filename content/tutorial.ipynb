# 🎓 Software Programming 4 Kids

## Interactive Learning Platform with Jupyter Lite

Welcome to an **interactive**, **offline-capable** learning platform where all code runs directly in your browser!

---

## ✨ What is JupyterLite?

- 🌐 **Browser-based** - No installation required
- 📱 **Offline-capable** - Works without internet
- ⚡ **Instant execution** - No server setup
- 🔒 **Safe** - All computation happens locally in your browser
- 📊 **Full Python stack** - NumPy, Pandas, Matplotlib, and more!

---

## 🔬 NumPy & Matplotlib Lessons

### **🚀 Physics**
- [**Projectile Motion**](numpy/physics/projectile_motion.ipynb) - Analyze trajectories and launch angles
  - Learn about kinematics, gravity, and range calculations
  - Experiment with different angles and velocities

- [**Atmospheric Pressure vs Altitude**](numpy/physics/atmospheric_pressure.ipynb) - Barometric formula
  - Understand how pressure decreases exponentially with altitude
  - Compare atmospheric layers (troposphere, stratosphere)
  - Real-world applications: Mount Everest, commercial flights

### **🧪 Chemistry**
- [**pH and H⁺ Ion Concentration**](numpy/chemistry/pH_concentration.ipynb) - Acid-base chemistry
  - Calculate hydrogen ion concentration using logarithmic relationships
  - Visualize acidic vs. alkaline regions
  - Explore real-world pH values

- [**Radioactive Decay: Half-Life**](numpy/chemistry/half_life.ipynb) - Nuclear chemistry
  - Analyze radioactive isotope decay over time
  - Compare Cobalt-60, Cesium-137, and Americium-241
  - View decay on linear and logarithmic scales

### **💰 Mathematics**
- [**Compound Interest Calculator**](numpy/math/compound_interest.ipynb) - Exponential growth
  - Explore how investments grow over time
  - Compare different interest rates and time periods
  - Learn the Rule of 72 and retirement planning

## 🚀 How to Use This Platform

### **Step 1: Open a Lesson**

Each lesson includes:
