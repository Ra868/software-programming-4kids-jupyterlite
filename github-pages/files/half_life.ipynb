````python3
print("PROGRAM NAME: Radioactive Decay & Half-Life Analysis\n")

import numpy as np
import matplotlib.pyplot as plt

# ==============================================
# EXPONENTIAL DECAY FORMULA
# ==============================================
# N(t) = N₀ × (0.5)^(t / T_half)
#
# Where:
# N(t) = Amount remaining at time t
# N₀ = Initial amount
# t = Time elapsed
# T_half = Half-life of the isotope
#
# Key Concept: Half-life is the time required for
# half of the radioactive atoms to decay.

# ==============================================
# SETUP
# ==============================================
N0 = 100  # Initial amount (grams)
time_years = np.array([1, 30, 50])  # Time points (years)

# Dictionary of isotopes with their half-lives (years)
# These are real radioactive isotopes used in industry, medicine, and research
half_lives = {
    "Cobalt-60 (5.3 years)": 5.3,
    "Cesium-137 (30.17 years)": 30.17,
    "Americium-241 (432 years)": 432
}

# ==============================================
# DECAY CALCULATION FUNCTION
# ==============================================
def remaining_amount(N0, t, T_half):
    """
    Calculate remaining amount of radioactive material after time t.
    
    Parameters:
    - N0: Initial amount (grams)
    - t: Time array (years)
    - T_half: Half-life (years)
    
    Returns:
    - Remaining amount(s) at time(s) t
    """
    return N0 * (0.5) ** (t / T_half)

# ==============================================
# COMPUTE RESULTS
# ==============================================
results = {}
for isotope, T_half in half_lives.items():
    results[isotope] = remaining_amount(N0, time_years, T_half)

# ==============================================
# DISPLAY TABLE
# ==============================================
print("Remaining Amount of Radioactive Isotopes")
print("Initial Amount: 100 grams\n")
print("-" * 75)
print(f"{'Isotope':35s} {'At 1 year':>15s} {'At 30 years':>15s} {'At 50 years':>15s}")
print("-" * 75)

for isotope, values in results.items():
    print(f"{isotope:35s} {values[0]:15.4f}g {values[1]:15.4f}g {values[2]:15.4f}g")
    # Also show percentage remaining
    percent = (values / N0) * 100
    print(f"{'':35s} ({percent[0]:6.2f}%) ({percent[1]:6.2f}%) ({percent[2]:6.2f}%)")
    print()

print("-" * 75)
print()
````

## 📊 Linear Scale Visualization

This plot shows actual amounts - notice how quickly some isotopes decay!

````python3
# ===== LINEAR SCALE PLOT =====
fig, ax = plt.subplots(figsize=(10, 6))

colors = ['#ff6b6b', '#4ecdc4', '#95e1d3']
for idx, (isotope, values) in enumerate(results.items()):
    ax.plot(time_years, values, marker='o', linewidth=2.5, 
            markersize=8, label=isotope, color=colors[idx], alpha=0.8)

# Formatting
ax.set_xlabel("Time (years)", fontsize=12, fontweight='bold')
ax.set_ylabel("Remaining Amount (grams)", fontsize=12, fontweight='bold')
ax.set_title("Radioactive Decay of Isotopes (Linear Scale)", 
             fontsize=14, fontweight='bold', pad=20)
ax.legend(fontsize=11, loc='upper right', framealpha=0.95)
ax.grid(True, alpha=0.3, linestyle='--')

# Add annotation
ax.text(0.5, 0.97, 'Notice: Cobalt-60 decays much faster than Americium-241', 
        transform=ax.transAxes, fontsize=10, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3))

plt.tight_layout()
plt.show()
````

## 📊 Logarithmic Scale Visualization

The log scale reveals the **exponential nature** of radioactive decay - all isotopes follow straight lines!

````python3
# ===== LOGARITHMIC SCALE PLOT =====
fig, ax = plt.subplots(figsize=(10, 6))

colors = ['#ff6b6b', '#4ecdc4', '#95e1d3']
for idx, (isotope, values) in enumerate(results.items()):
    ax.plot(time_years, values, marker='o', linewidth=2.5, 
            markersize=8, label=isotope, color=colors[idx], alpha=0.8)

# Use logarithmic scale for y-axis
ax.set_yscale('log')

# Formatting
ax.set_xlabel("Time (years)", fontsize=12, fontweight='bold')
ax.set_ylabel("Remaining Amount (grams) [Log Scale]", fontsize=12, fontweight='bold')
ax.set_title("Radioactive Decay of Isotopes (Logarithmic Scale)", 
             fontsize=14, fontweight='bold', pad=20)
ax.legend(fontsize=11, loc='upper right', framealpha=0.95)
ax.grid(True, which='both', alpha=0.3, linestyle='--')

# Add annotation
ax.text(0.5, 0.05, 'On a log scale, exponential decay appears as straight lines!', 
        transform=ax.transAxes, fontsize=10, verticalalignment='bottom',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.3))

plt.tight_layout()
plt.show()
````

## 🧪 Key Chemistry & Physics Concepts

### **Half-Life Definition**
Half-life (T₁/₂) is the time required for half of the atoms in a radioactive sample to decay.

### **Examples from this program:**
- **Cobalt-60**: T₁/₂ = 5.3 years
  - Used in medical treatments and industrial applications
  - After 5.3 years: 50g remains
  - After 10.6 years: 25g remains

- **Cesium-137**: T₁/₂ = 30.17 years
  - Produced in nuclear reactors and weapons
  - Significant environmental contaminant
  - After 30.17 years: 50g remains

- **Americium-241**: T₁/₂ = 432 years
  - Used in smoke detectors
  - Poses long-term environmental hazard
  - After 432 years: still 50g remains!

### **Mathematical Pattern**
Notice the exponential decay pattern:
- After 1 half-life: 50% remains
- After 2 half-lives: 25% remains
- After 3 half-lives: 12.5% remains
- After n half-lives: (0.5)^n % remains

### **Why Two Scales?**
- **Linear scale** shows actual amounts (intuitive)
- **Log scale** reveals the exponential relationship (mathematical insight)
